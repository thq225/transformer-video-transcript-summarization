```python
from transformers import BartTokenizerFast, BartForConditionalGeneration

MODEL_DIR = "../outputs/qmsum_bart_baseline/best_model"

tokenizer = BartTokenizerFast.from_pretrained(MODEL_DIR)
model = BartForConditionalGeneration.from_pretrained(MODEL_DIR)
```

```python
query = "What decisions were made about the interface?"

transcript = """
Speaker A: We need to redesign the interface because users find it confusing.
Speaker B: I think simplifying the navigation bar would help.
Speaker C: Yes, and we should also improve the search functionality.
Speaker A: So the decision is to simplify the navigation and improve search.
"""

input_text = f"query: {query}\n\ntranscript:\n{transcript}"
print(input_text)
```

```python
inputs = tokenizer(
    input_text,
    return_tensors="pt",
    max_length=1024,
    truncation=True,
)
```

```python
summary_ids = model.generate(
    **inputs,
    max_length=128,
    num_beams=4,
    early_stopping=True,
)
```

```python
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Generated summary:\n")
print(summary)
```

```python
query = "What are the main concepts explained in the lecture?"

lecture_transcript = """
Professor: Today we will discuss the basics of neural networks. A neural network is a model inspired by the human brain.

Professor: The key building block of neural networks is the neuron. Each neuron receives inputs, multiplies them by weights, sums them, and passes the result through an activation function.

Professor: Common activation functions include ReLU, sigmoid, and tanh.

Professor: Neural networks are trained using gradient descent. During training, we compute the loss between predicted outputs and true labels.

Professor: Then we use backpropagation to update the weights in order to reduce the loss.

Professor: In deep learning, we stack many layers of neurons to create deep neural networks. These models have been very successful in tasks such as image recognition, speech recognition, and natural language processing.

Professor: However, deep networks require large datasets and significant computational power to train effectively.
"""

input_text = f"query: {query}\n\ntranscript:\n{lecture_transcript}"

inputs = tokenizer(
    input_text,
    return_tensors="pt",
    max_length=1024,
    truncation=True
)

summary_ids = model.generate(
    **inputs,
    max_length=128,
    num_beams=4,
    early_stopping=True
)

summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Generated summary:\n")
print(summary)
```
